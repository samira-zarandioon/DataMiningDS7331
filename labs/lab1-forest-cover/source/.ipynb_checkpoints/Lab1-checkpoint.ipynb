The Roosevelt National Forest is a National Forest located in north central Colorado with a total area of 813,799 acres.  The forest is administered jointly with the Arapaho National Forest and the Pawnee National Grassland.  In order to maintain and improve the health, diversity, and productivity of the forest, the natural resource managers are responsible for developing ecosystem management strategies. Developing these strategies require information and data that the forest managers generally do not have.  To solve for the lack of information, this data set was collected to build models to predict forest cover types that would help forest managers maintain forest health. 

The data collected was derived from data originally obtained from US Geological Survey (USGS) and USFS data.  The data comes from wilderness areas located within the Roosevelt National Forest that have had minimal human interactions, so we can be more confident that the current forest cover type is more of a result of natural ecological processes rather than forest management practices.

Results from any prediction algorithm developed using this data can be measured against the actual forest cover type determined from US Forest Service (USFS) Region 2 Resource Information System (RIS) data.
